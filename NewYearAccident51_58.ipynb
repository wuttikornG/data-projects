```python
# clean data ( day can not equal 0)
df = df[df['วันที่เกิดเหตุ']!=0]
```

```python
df['ชื่อเทศกาล'].str[-2:]
```

```python
# change format column ชื่อเทศกาล 'ปีใหม่ 51' to '2008' //convert พ.ศ. to ค.ศ.
df['ปีที่เกิดเหตุ'] = pd.DataFrame('25'+df['ชื่อเทศกาล'].str[-2:]).astype('int64')-543
df['ปีที่เกิดเหตุ']
```

```python
# def fuction to define 'month'
def map_month(day):
    if 1 <= day <= 5:
        return 1
    elif 27 <= day <= 31:
        return 12
    else:
        return None
```

```python
df['เดือนที่เกิดเหตุ'] = df['วันที่เกิดเหตุ'].apply(map_month)
```

```python
df['date'] = pd.to_datetime(dict(year=df['ปีที่เกิดเหตุ'], month=df['เดือนที่เกิดเหตุ'], day=df['วันที่เกิดเหตุ']), format='%Y%m%d')
df.head()
```

Find the number of the dead cases in each province and vehicle types.

```python
df['ผลการรักษา'].unique()
```

```python
# filter df only dead cases
dead_df = df[df['ผลการรักษา'] != 'ทุเลา/หาย']
```

```python
# pivot table to show the number of the dead cases in each province and vehicle types
pivot_dead_df = dead_df.pivot_table(index = 'จังหวัด', columns = 'รถผู้บาดเจ็บ', values = 'ผลการรักษา', aggfunc='count').fillna(0)
pivot_dead_df
```

What day are the most accident cases reported?

```python
df['วันที่เกิดเหตุ'].value_counts()
```

```python
# Extract 'day-month' from the 'date' column
df['day_month'] = df['date'].dt.strftime('%d-%m')

# Count accidents of each 'day-month'
count_per_day_month = df['day_month'].value_counts().sort_index()

# set the order of x-axis (Dec first, then Jan)
custom_order = [f'{day:02d}-12' for day in range(27, 32)] + \
               [f'{day:02d}-01' for day in range(1, 6)]
```

```python
import matplotlib.pyplot as plt

# Plot Bar Chart
plt.figure(figsize=(10, 6))
count_per_day_month.reindex(custom_order).plot(kind='bar', color='skyblue')
plt.title('Count of accidents reported per day-month')
plt.xlabel('Day-Month')
plt.ylabel('Count')
plt.show()
```

Find the accident counts by Top 5 vehicle types and Top 10 provinces.

```python
top_provinces = df['จังหวัด'].value_counts().head(10)
top_provinces
```

```python
# Plot Bar Chart
plt.figure(figsize=(10, 6))
top_provinces.plot(kind='bar', color='skyblue')
plt.xticks(rotation=0)
plt.title('Top 10 Provinces')
plt.xlabel('จังหวัด')
plt.ylabel('Count')

for bar, count in zip(plt.gca().patches, top_provinces):
    plt.annotate(str(count), xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='bottom', fontsize=10)
plt.show()
```

```python
import seaborn as sns
# top 10 provinces
top_provinces_index = df['จังหวัด'].value_counts().head(10).index

# top 5 vehicle types
top_vehicles_index = df['รถผู้บาดเจ็บ'].value_counts().head().index

# filter Top 10 provinces and Top 5 vehicle types
filtered_df = df[(df['จังหวัด'].isin(top_provinces_index)) & (df['รถผู้บาดเจ็บ'].isin(top_vehicles_index))]

# use seaborn heatmap to show color table
pivot_veh_prov = pd.pivot_table(filtered_df, values='date', index='จังหวัด', columns='รถผู้บาดเจ็บ', aggfunc='count', fill_value=0)
sns.heatmap(pivot_veh_prov, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Accident Counts by Top 5 Vehicles and Top 10 Provinces')
plt.xlabel('Vehicle Type')
plt.ylabel('Province')
plt.show()
```
