# API Thai Postal Codes

```python
# import
import requests
import pandas as pd
```

```python
# Define the range of postal codes
start_postal_code = 10100
end_postal_code = 10700

# Initialize a list to store the postal code information
postal_code_data = []

# API URL for the range of postal codes (skip range every 10 postal codes)
for postal_code in range(start_postal_code, end_postal_code + 1, 10):

    api_url = f"https://api.zippopotam.us/th/{postal_code}"

    # GET request to the API
    response = requests.get(api_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON response to extract postal code information
        data = response.json()
        if 'places' in data:
            place_data = data['places'][0]
            # Append the place data to the list
            postal_code_data.append({
                'postcode': postal_code,
                'place name': place_data.get('place name', ''),
                'state': place_data.get('state', ''),
                'state abbreviation': place_data.get('state abbreviation', ''),
                'longitude': place_data.get('longitude', ''),
                'latitude': place_data.get('latitude', '')
            })
    # Skip collecting data when status code is not 200
    else:
        print(f"Skipped postal code: {postal_code} (Status Code: {response.status_code})")
```

```python
# Create a DataFrame from the collected postal code information with the desired column order
df = pd.DataFrame(postal_code_data, columns=['postcode', 'place name', 'state', 'state abbreviation', 'longitude', 'latitude'])
df.head(10)
```

```python
# Write the csv file
df.to_csv("Thai_postal_codes.csv")
```
