```python
import pandas as pd
```

Read the data file and store in a variable.

```python
df = pd.read_csv("confirmed-cases-since-120864.csv")
df
```

```python
df.dtypes
```

Convert the 'announce_date' series to be datetime (use the dayfirst parameter).

```python
df['announce_date'] = pd.to_datetime(df['announce_date'], dayfirst=True)
```

```python
df['announce_date'].value_counts()
```

```python
df.head()
```

```python
df['province_of_isolation'].value_counts().index
```

```python
df[df['province_of_isolation']=='กทม']
```

Replace 'กทม' with 'กรุงเทพมหานคร'

```python
df['province_of_isolation'] = df['province_of_isolation'].replace('กทม', 'กรุงเทพมหานคร')
```

```python
df['province_of_isolation'].value_counts()
```

Use groupby with a list of series, ['province_of_isolation','announce_date'], as the parameter. To count all 'No.' series on each day of each province.

```python
df.groupby(['province_of_isolation', 'announce_date'])['No.'].count()
```

```python
df.groupby(['province_of_isolation', 'announce_date'])['No.'].count().reset_index()
```

Create the dataframe as describe above using the pivot function.

```python
result_df = df.pivot_table(index = 'announce_date', columns = 'province_of_isolation', values = 'No.', aggfunc='count')
result_df
```

Check NA and replace it with 0

```python
result_df.isna().sum().sort_values(ascending = False)
```

```python
result_df.fillna(0,inplace=True)
```

```python
result_df.columns
```

QUESTION : จังหวัดไหนที่มีแนวโน้มพบผู้ติดเชื้อโควิดใกล้เคียงกันในช่วงเวลานั้นๆ

Create list of sorted distances(number of confirmed cases) between all pairs of province

```python
cols=result_df.columns
distances=[]
for i in range(len(cols)-1):
  for j in range(i+1,len(cols)):
    distances.append((sum((result_df[cols[i]]-result_df[cols[j]])**2),cols[i],cols[j]))
distances.sort()
```

```python
distances[:5]
```

```python
result_df['มุกดาหาร'].plot()
result_df['แพร่'].plot()
```

```python
distances[-5:]
```

```python
result_df['กรุงเทพมหานคร'].plot()
result_df['แม่ฮ่องสอน'].plot()
```

Find the top 5 pairs of provinces with the smallest distances between them.

```python
import matplotlib.pyplot as plt

for t in distances[:5]:
  print(t[1],t[2])
  result_df[t[1]].plot()
  result_df[t[2]].plot()
  plt.show()
```
